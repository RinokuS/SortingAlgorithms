```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_big = pd.read_csv("big.csv", sep=';')
data_small = pd.read_csv("small.csv", sep=';')
```

# Генерация графиков по видам массивов

```python
# разбиваем на две группы, ибо вторая вырождается в ось x на одном графике с первой
types_of_sorts_first = ['Bubble', 'Selection', 'Insertion', 'Merge']
types_of_sorts_second = ['Counting', 'Radix', 'Quick', 'Heap']

types_of_arrs = ['Random 0-5', 'Random 0-4000', 'Almost Sorted', 'Reversed Sorted']
```

## Малые массивы

### Первая группа

```python
fig, axes = plt.subplots(4)
i = 0

for ax in axes.ravel():
    for sort in types_of_sorts_first:
        ax.plot(data_small['Unnamed: 0'], data_small[sort + ' ' + types_of_arrs[i]], label=sort)
    ax.legend()
    ax.set_title(types_of_arrs[i])
    ax.set_xlabel('number of elements')
    ax.set_ylabel('time (ticks)')
    i += 1

fig.set_figheight(30)
fig.set_figwidth(15)
```

### Вторая группа

```python
fig, axes = plt.subplots(4)
i = 0

for ax in axes.ravel():
    for sort in types_of_sorts_second:
        ax.plot(data_small['Unnamed: 0'], data_small[sort + ' ' + types_of_arrs[i]], label=sort)
    ax.legend()
    ax.set_title(types_of_arrs[i])
    ax.set_xlabel('number of elements')
    ax.set_ylabel('time (ticks)')
    i += 1

fig.set_figheight(30)
fig.set_figwidth(15)
```
